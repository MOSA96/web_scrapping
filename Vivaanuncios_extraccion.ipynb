```python
#Importamos librerías

import bs4
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq
from urllib.request import Request
import time
import pandas as pd
```

```python
#Creamos listas vacias
cuartos = []
banos = []
area = []
colonia = []
precios2 = []

#Extraemos datos

for i in range(1,51):

    my_url = "https://www.vivanuncios.com.mx/s-venta-inmuebles/distrito-federal/page-{}/v1c1097l1008p{}".format(i,i)

    uClient = uReq(my_url)
    page_html = uClient.read()
    uClient.close()

    page_soup = soup(page_html, "html.parser")
    
    
    precios = page_soup.findAll("span", {"class":"ad-price"})
    precios = precios[10::]

    features = page_soup.findAll("div", {"class":"float-right"})
    lugar = page_soup.findAll("div", {"class":"tile-location one-liner"})
    descripcion = page_soup.findAll("a", {"class":"href-link tile-title-text"})

    for each in features:
        try:
            cuartos.append((each.find("div", {"class":"chiplets-inline-block re-bedroom"}).text))
        except:
            cuartos.append(None)
        
    for each in features:
        try:
            banos.append((each.find("div", {"class":"chiplets-inline-block re-bathroom"}).text))
        except:
            banos.append(None)
               
    for each in features:
        try:
            area.append((each.find("div", {"class":"chiplets-inline-block surface-area"}).text))
        except:
            area.append(None)
        
    for each in lugar:
        try:
            colonia.append(each.b.string)
        except:
            colonia.append(None)

    for i in range(len(precios)):
        precios2.append(precios[i].text.strip())
        
```

```python
#Creamos dataframe

names = ["precio", "cuartos", "baños", "area", "colonia"]

df = pd.DataFrame(columns=names)
df["precio"] = precios2
df["cuartos"] = cuartos
df["baños"] = banos
df["area"] = area
df["colonia"] = colonia
```

```python
#Eliminamos datos con valores NaN

df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
```
